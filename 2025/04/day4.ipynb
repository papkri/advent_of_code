## Day 4
### Part 1

```python
import numpy as np
def part1(map_layout):
    windows = np.lib.stride_tricks.sliding_window_view(map_layout, (3, 3))

    # compute center indices
    H, W = map_layout.shape
    centers = np.stack(np.mgrid[1:H-1, 1:W-1], axis=-1)

    # flatten both
    windows_flat = windows.reshape(-1, 3, 3)
    centers_flat = list(map(tuple, centers.reshape(-1, 2)))
    map_layout_copy = map_layout.copy()

    for window, center in zip(windows_flat, centers_flat):
        if map_layout_copy[center] == ".":
            continue

        if np.sum(window == "@") <= 4:
            map_layout_copy[center] = "x"
        else:
            continue

    return(np.sum(map_layout_copy == "x"))
```

```python

map_layout = np.genfromtxt("day4.txt", dtype = "U1", delimiter = 1)  
map_layout = np.pad(map_layout, pad_width=1, mode = "constant", constant_values=".")
part1(map_layout)
```

### Part 3

```python
def part2(map_layout):

    something_left_to_change = True
    while something_left_to_change:
        map_at_beginning = map_layout.copy()

        windows = np.lib.stride_tricks.sliding_window_view(map_layout, (3, 3))
        H, W = map_layout.shape
        centers = np.stack(np.mgrid[1:H-1, 1:W-1], axis=-1)

        # flatten both
        windows_flat = windows.reshape(-1, 3, 3)
        centers_flat = list(map(tuple, centers.reshape(-1, 2)))

        for window, center in zip(windows_flat, centers_flat):
            if map_layout[center] == ".":
                continue

            if np.sum(window == "@") <= 4:
                map_layout[center] = "x"
            else:
                continue

        if np.array_equal(map_at_beginning, map_layout):
            print("testoid")
            something_left_to_change = False

    return(np.sum(map_layout == "x"))
```

```python
map_layout = np.genfromtxt("day4.txt", dtype = "U1", delimiter = 1)  
map_layout = np.pad(map_layout, pad_width=1, mode = "constant", constant_values=".")
part2(map_layout)
```
