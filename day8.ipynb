```python
import numpy as np
def advent8_part1(map_layout):

    unique_value_counts = np.unique_counts(map_layout)
    mask = np.all([unique_value_counts.values != ".", unique_value_counts.counts >= 2], axis = 0)
    valid_antennas = unique_value_counts.values[mask]

    antinodes = map_layout.copy()
    print(antinodes[11,9])

    for valid_antenna in valid_antennas:
        indices = np.argwhere(map_layout == valid_antenna)
        pairings_iterator = combinations(indices, 2)
        pairings_iterator = np.array(list(pairings_iterator))

        for pairing in pairings_iterator:
            a_x = pairing[0][0]
            a_y = pairing[0][1]
            b_x = pairing[1][0]
            b_y = pairing[1][1]

            new_node_a = np.array([2*a_x - b_x, 2*a_y - b_y])
            new_node_b = np.array([2*b_x - a_x, 2*b_y - a_y])

            for new_node in [new_node_a, new_node_b]:

                if any([new_node[0] < 0, new_node[0] > (antinodes.shape[0]-1), new_node[1] < 0, new_node[1] > (antinodes.shape[1]-1)]):
                    continue
                else:
                    if antinodes[new_node[0], new_node[1]] != "#":
                        antinodes[new_node[0], new_node[1]] = "#"

    return np.sum(antinodes == "#")
```

```python
map_layout = np.genfromtxt('day8.txt', dtype='U1', delimiter=1,
                           comments = None)
advent8_part1(map_layout)
```
