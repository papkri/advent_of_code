```python
def part1(numbers, operations):
    rows, cols = numbers.shape
    runsum = 0
    for j in range(cols):
        col_data = numbers[:, j]
        op = operations[j]
        if op == "+":
            runsum += np.sum(col_data)
        elif op == "*":
            runsum += np.prod(col_data)

    return runsum

part1(numbers, operations)
```

```python
def get_character_at_index(numpy_string_array, index):
    """Return a 1D array of characters taken from each string in
    `numpy_string_array` at position `index`. If the index is negative,
    we count from the right (so -1 is last char). If a string is too
    short, an empty string '' is returned for that element."""
    def char_extractor(s):
        if index < 0:
            abs_index = abs(index)
            return s[-abs_index] if len(s) >= abs_index else ''
        else:
            return s[index] if len(s) > index else ''

    vectorized_extractor = np.vectorize(char_extractor, otypes=[str])
    return vectorized_extractor(numpy_string_array)


def transform_column(col_data):
    """Given a 1D integer array `col_data` (top-to-bottom), build an
    array of integers formed by concatenating the characters taken
    from each row at each digit position (left-to-right). Missing digits
    are skipped so the result can be shorter when leading rows lack that digit.

    Return order: rightmost column first, then moving left (so for
    [64,23,314] -> [4, 431, 623]).
    """
    maximum = np.max(col_data)
    len_max = len(str(maximum))
    str_array = col_data.astype(str)
    transformed_column = []

    # iterate column indices from rightmost (len_max-1) down to 0
    for i in range(len_max - 1, -1, -1):
        chars = get_character_at_index(str_array, i)
        joined = ''.join(chars)
        transformed_column.append(int(joined))

    return np.array(transformed_column)
# ...existing code...


def part2(numbers, operations):
    rows, cols = numbers.shape
    runsum = 0
    for j in range(cols):
        col_data = transform_column(numbers[:, j])
        print(f"Column {j} transformed -> {col_data}")
        op = operations[j]
        if op == "+":
            runsum += np.sum(col_data)
        elif op == "*":
            runsum += np.prod(col_data)

    return runsum

# Demo: show results for current `numbers` and `operations`
print("part2:", part2(numbers, operations))
```
