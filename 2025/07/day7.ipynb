## Day 7
### Part 1

```python
def move_beam(coord_x, coord_y, manifold):
    if manifold[coord_x, coord_y] == ".":
        return None
    elif manifold[coord_x, coord_y] == "S" or manifold[coord_x, coord_y] == "|":
        if manifold[coord_x+1, coord_y] == ".":
            manifold[coord_x+1, coord_y] = "|"
        elif manifold[coord_x+1, coord_y] == "^":
            manifold[coord_x+1, coord_y-1] = "|"
            manifold[coord_x+1, coord_y+1] = "|"
            return 1

    return 0
```

```python
manifold = np.genfromtxt("day7.txt", dtype=str, delimiter = 1)
runsum = 0
for i in range(manifold.shape[0]-1):
    for j in range(manifold.shape[1]):
        output = move_beam(i, j, manifold)
        if output == 1:
            runsum += 1
print(runsum)
```

### Part 2

```python
def move_beam_part2(coord_x, coord_y, manifold, results_manifold):
    if manifold[coord_x, coord_y] == ".":
        return None
    elif manifold[coord_x, coord_y] == "S" or manifold[coord_x, coord_y] == "|":
        if manifold[coord_x+1, coord_y] == "." or manifold[coord_x+1, coord_y] == "|":
            manifold[coord_x+1, coord_y] = "|"
            results_manifold[coord_x+1, coord_y] += results_manifold[coord_x, coord_y]
        elif manifold[coord_x+1, coord_y] == "^":
            manifold[coord_x+1, coord_y-1] = "|"
            results_manifold[coord_x+1, coord_y-1] +=  results_manifold[coord_x, coord_y]
            manifold[coord_x+1, coord_y+1] = "|"
            results_manifold[coord_x+1, coord_y+1] +=  results_manifold[coord_x, coord_y]
            return 1

    return 0

manifold = np.genfromtxt("day7.txt", dtype=str, delimiter = 1)
results_manifold = np.zeros(manifold.shape)
results_manifold[np.where(manifold == "S")] = 1
for i in range(manifold.shape[0]-1):
    for j in range(manifold.shape[1]):
        output = move_beam_part2(i, j, manifold, results_manifold)

np.sum(results_manifold[-1])
```
