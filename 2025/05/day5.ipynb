```python
file_path = 'day5_ranges.txt' # Make sure this is the actual path to your file
ranges_list = []

with open(file_path, 'r') as file:
    for line in file:
        # Strip newline characters and split the line by the hyphen
        parts = line.strip().split('-')
        # Convert the resulting strings into integers and store them as a tuple/list
        if len(parts) == 2:
            ranges_list.append((int(parts[0]), int(parts[1])))

# 'data' now contains a list of tuples, e.g., [(3, 5), (10, 14), (16, 20), (12, 18)]
print(ranges_list)
```

```python
import numpy as np
ids = np.loadtxt("day5_ids.txt", dtype=int)
```

```python
is_in_any_range = np.zeros(ids.shape, dtype=bool)

for start, end in ranges_list:
    in_current_range = (ids >= start) & (ids <= end)
    is_in_any_range = is_in_any_range | in_current_range

print(sum(is_in_any_range))

```

```python
def calculate_overlap_size(range_a, range_b):
    # range_a and range_b are tuples or arrays like (start, end)
    overlap_start = np.maximum(range_a[0], range_b[0])
    overlap_end = np.minimum(range_a[1], range_b[1])
    # Calculate length, ensuring it's not negative
    overlap_length = np.maximum(0, overlap_end - overlap_start + 1)
    return overlap_length
```

```python
num_ranges = len(ranges_list)
# Initialize an empty matrix to store results
overlap_matrix = np.zeros((num_ranges, num_ranges), dtype=int)

for i in range(num_ranges):
    for j in range(num_ranges):
        # Fill the matrix cell [i, j] with the overlap size of range i and range j
        overlap_matrix[i, j] = calculate_overlap_size(ranges_list[i], ranges_list[j])
```
