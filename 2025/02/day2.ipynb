## Part 1

```python
import re

def is_id_invalid(num_string):
    midpoint = len(num_string) // 2
    first_half = num_string[:midpoint]
    second_half = num_string[midpoint:]

    if first_half == second_half:
        return True
    else:
        return False


def process_id_range(id_range):
    pattern = r"^(\d+)-(\d+)$"
    match = re.search(pattern, id_range)
    
    start_num = int(match.group(1))
    end_num = int(match.group(2))

    runsum = 0
    for i in range(start_num, end_num+1):
        if is_id_invalid(str(i)):
            runsum += i
            
    return (runsum)
```

```python
file_name = 'day2.txt'

with open(file_name, 'r') as file:
    first_line = file.readline()
    cleaned_line = first_line.strip()
    id_list = [item.strip() for item in cleaned_line.split(',')]

runsum = 0
for id_range in id_list:
    runsum += process_id_range(id_range)
print(runsum)
```

## Part 2

```python
import re

def is_id_invalid_part2(num_string):

    for id in range(1, (len(num_string) // 2)+1):
        string_component = num_string[:id]
        regex_pattern = f"^(?:{string_component})+$"
        if re.fullmatch(regex_pattern, num_string):
            return True
        else:
            continue

    return False


def process_id_range_part2(id_range):
    pattern = r"^(\d+)-(\d+)$"
    match = re.search(pattern, id_range)
    
    start_num = int(match.group(1))
    end_num = int(match.group(2))

    runsum = 0
    for i in range(start_num, end_num+1):
        if is_id_invalid_part2(str(i)):
            runsum += i
            
    return (runsum)
```

```python
runsum = 0
for id_range in id_list:
    runsum += process_id_range_part2(id_range)
print(runsum)
```
