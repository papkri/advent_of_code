# Advent of Code Day 8
## Part 1

```python
import numpy as np
import itertools
def advent8_part1(map_layout):

    unique_value_counts = np.unique_counts(map_layout)
    mask = np.all([unique_value_counts.values != ".", unique_value_counts.counts >= 2], axis = 0)
    valid_antennas = unique_value_counts.values[mask]

    antinodes = map_layout.copy()
    print(antinodes[11,9])

    for valid_antenna in valid_antennas:
        indices = np.argwhere(map_layout == valid_antenna)
        pairings_iterator = combinations(indices, 2)
        pairings_iterator = np.array(list(pairings_iterator))

        for pairing in pairings_iterator:
            a_x = pairing[0][0]
            a_y = pairing[0][1]
            b_x = pairing[1][0]
            b_y = pairing[1][1]

            new_node_a = np.array([2*a_x - b_x, 2*a_y - b_y])
            new_node_b = np.array([2*b_x - a_x, 2*b_y - a_y])

            for new_node in [new_node_a, new_node_b]:

                if any([new_node[0] < 0, new_node[0] > (antinodes.shape[0]-1), new_node[1] < 0, new_node[1] > (antinodes.shape[1]-1)]):
                    continue
                else:
                    if antinodes[new_node[0], new_node[1]] != "#":
                        antinodes[new_node[0], new_node[1]] = "#"

    return np.sum(antinodes == "#")
```

```python
map_layout = np.genfromtxt('day8.txt', dtype='U1', delimiter=1,
                           comments = None)
advent8_part1(map_layout)
```

## Part 2

```python
import numpy as np
from itertools import combinations
def advent8_part2(map_layout):

    unique_value_counts = np.unique_counts(map_layout)
    mask = np.all([unique_value_counts.values != ".", unique_value_counts.counts >= 2], axis = 0)
    valid_antennas = unique_value_counts.values[mask]

    antinodes = map_layout.copy()
    n_rows, n_cols = antinodes.shape[0], antinodes.shape[1]
    print(antinodes[11,9])

    for valid_antenna in valid_antennas:
        indices = np.argwhere(map_layout == valid_antenna)
        pairings_iterator = combinations(indices, 2)
        pairings_iterator = np.array(list(pairings_iterator))

        for pairing in pairings_iterator:
            a_x = pairing[0][0]
            a_y = pairing[0][1]
            b_x = pairing[1][0]
            b_y = pairing[1][1]

            new_nodes = []
            #extend in direction_a until we are out of the map
            while True:
                new_node_a = np.array([2*a_x - b_x, 2*a_y - b_y])
                if any([new_node_a[0] < 0, new_node_a[0] > (n_rows-1), new_node_a[1] < 0, new_node_a[1] > (n_cols-1)]):
                    break
                else:
                    new_nodes.append(new_node_a)
                    b_x, b_y = a_x, a_y
                    a_x, a_y = new_node_a[0], new_node_a[1]
            
            #extend in direction_b until we are out of the map
            a_x = pairing[0][0]
            a_y = pairing[0][1]
            b_x = pairing[1][0]
            b_y = pairing[1][1]
            
            while True:
                new_node_b = np.array([2*b_x - a_x, 2*b_y - a_y])
                if any([new_node_b[0] < 0, new_node_b[0] > (n_rows-1), new_node_b[1] < 0, new_node_b[1] > (n_cols-1)]):
                    break
                else:
                    new_nodes.append(new_node_b)
                    a_x, a_y = b_x, b_y
                    b_x, b_y = new_node_b[0], new_node_b[1]

            #add the new nodes to the map
            for new_node in new_nodes:
                if antinodes[new_node[0], new_node[1]] != "#":
                    antinodes[new_node[0], new_node[1]] = "#"
                        
    #add the antinodes that appear in valid anteannas
    antinodes[np.isin(antinodes, valid_antennas)] = "#"

    return np.sum(antinodes == "#")
```

```python
map_layout = np.genfromtxt('day8.txt', dtype='U1', delimiter=1,
                           comments = None)
advent8_part2(map_layout)
```
