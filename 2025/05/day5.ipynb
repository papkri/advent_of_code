## Part 1

```python
file_path = 'day5_ranges.txt' # Make sure this is the actual path to your file
ranges_list = []

with open(file_path, 'r') as file:
    for line in file:
        # Strip newline characters and split the line by the hyphen
        parts = line.strip().split('-')
        # Convert the resulting strings into integers and store them as a tuple/list
        if len(parts) == 2:
            ranges_list.append((int(parts[0]), int(parts[1])))

import numpy as np
ids = np.loadtxt("day5_ids.txt", dtype=int)
```

For each range in the ranges list we will check whether any of the ids are within

```python
is_in_any_range = np.zeros(ids.shape, dtype=bool)

for start, end in ranges_list:
    in_current_range = (ids >= start) & (ids <= end)
    is_in_any_range = is_in_any_range | in_current_range

print(sum(is_in_any_range))
```

## Part 2
The arrays are too big to brute force it. Instead, we'll loop through the ranges as sorted by their start, and if the current range is bigger (its end is larger) then we increase the range otherwise if they are non-overlapping we createa  new range.

```python
ranges_list.sort()

merged = []
cur_start, cur_end = ranges_list[0]

for s, e in ranges_list[1:]:
    if s > cur_end:
        merged.append((cur_start, cur_end))
        cur_start, cur_end = s, e
    else:
        cur_end = max(cur_end, e)

merged.append((cur_start, cur_end))

# count unique values (exclusive end example)
unique = sum(e - s +1 for s, e in merged)
print(unique)
```
