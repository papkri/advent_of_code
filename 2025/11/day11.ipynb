```python
data_dict = {}

with open('day11.txt', mode='r', encoding='utf-8') as text_file:
    for line in text_file:
        if not line.strip():
            continue
        key, value_str = line.strip().split(':', 1)
        data_dict[key.strip()] = value_str.strip().split()
```

```python
def find_all_paths_dag(graph, start, end, path=[], paths=[]):
    path = path + [start]
    if start == end:
        paths.append(list(path)) # Base Case: Destination reached
        return
    if start not in graph:
        return # Base Case: Dead end

    for node in graph[start]:
        if node not in path: 
            find_all_paths_dag(graph, node, end, path, paths)

results = []
find_all_paths_dag(data_dict, start = "you", end = "out", paths = results)
```
