```python
import numpy as np
numbers = np.loadtxt('day6_numbers.txt', dtype=int)
```

```python
numbers
```

```python
operations = np.genfromtxt("day6_operations.txt", dtype=str)
```

```python
rows, cols = numbers.shape
runsum = 0
for j in range(cols):
    col_data = numbers[:, j]
    op = operations[j]
    if op == "+":
        runsum += np.sum(col_data)
    elif op == "*":
       runsum += np.prod(col_data)

runsum
```
